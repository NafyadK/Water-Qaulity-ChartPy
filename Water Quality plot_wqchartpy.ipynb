```python
# Import WQChartPy
from wqchartpy import triangle_piper
fig = plt.figure(figsize=(4, 3))
dataset['Marker'] = '^'
dataset['Size'] = 21
dataset['Alpha'] = 0.6

# Draw the diagram
triangle_piper.plot(dataset, unit='mg/L', figname='triangle Piper diagram', figformat='jpg')
```

```python
from wqchartpy import gibbs

gibbs.plot(dataset, unit='mg/L', figname='Gibbs diagram', figformat='jpg')
```
